```python
import fastf1
import datetime
```

```python
from fastf1.ergast import Ergast
from fastf1.ergast.interface import ErgastRawResponse

ergast = Ergast()
```

```python
ConstructorToID = dict()
```

```python
drivers = ergast.get_driver_info(season=2021, result_type='raw')
for d in drivers:
    firstName = d.get('givenName')
    lastName = d.get('familyName')
    dID = d.get('permanentNumber')
    code = d.get('code')
    country = d.get('nationality')
    age = int((datetime.datetime.now() - d.get('dateOfBirth')).days / 365)
    print(f"({dID}, \'{firstName}\', \'{lastName}\', \'{code}\', \'{country}\', {age}),")
```

```python
constructors = ergast.get_constructor_info(season=2021, result_type='raw')
index = 1
for c in constructors:
    name = c.get('name')
    constructorId = c.get('constructorId')
    print(f"({index}, \'{name}\'),")
    ConstructorToID[constructorId] = index
    index += 1
```

```python
schedule = ergast.get_race_schedule(season=2024)
num_races = len(schedule)
```

```python
for i in range(1, num_races + 1):
    results = ergast.get_race_results(season=2021, round=i, result_type='raw')
    for r in results:
        name = r.get("Circuit").get("circuitName")
        country = r.get("Circuit").get("Location").get("country")
        print(f"({i}, \'{name}\', \'{country}\', {i}, 2021),")
    
```

```python
results = ergast.get_race_results(season=2021, round=1, result_type='raw')

for r in results[0].get("Results"):
    position = r.get("position")
    points = r.get("points")
    print(f"(2021, {position}, {points}),")
```

```python
for i in range(1, num_races + 1):
    results = ergast.get_race_results(season=2021, round=i, result_type='raw')
    for result in results:
        for r in result.get('Results'):
            dID = r.get('Driver').get('permanentNumber')
            cID = ConstructorToID[r.get('Constructor').get('constructorId')]
            startPos = r.get('grid')
            finishPos = r.get('position')
            print(f"({dID}, {cID}, {i}, {startPos}, {finishPos}),")
```
